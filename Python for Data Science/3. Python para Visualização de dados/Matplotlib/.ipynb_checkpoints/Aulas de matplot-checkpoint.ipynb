```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0 , 5, 11)
x
```

```python
y =x * x
y
```

```python
#funcional
plt.plot(x, y)
# quando não utilizamos o comando %matplotlib inline
#temos que utilizar plt.show() para que o output seja um gráfico plotado
```

```python
#Adicionando descritores de título do grafico eixo x e y
plt.title('Meu primeiro gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
#linha do gráfico na cor vermelha e tracejada
plt.plot(x, y, 'r--')
```

```python
plt.subplot(1, 2, 1) #com decrição dos argumentos
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2) #Com resumo dos argumentos
plt.plot(y, x, 'g*-') #linha verde tracejada marcad com estrelas
```

```python
#criando figuras
fig = plt.figure()
#gerando uma figura em dada posição
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#os argumentos são: 
# 1- percentual de posicionamento em relação a esquedra
# 2- percentual de posicionamento em relação a parte de baixo
# 3- Largura
# 4- Altura
axes.plot(x, y)
axes.set_xlabel('Eixo X')
axes.set_ylabel('Eixo Y')
axes.set_title("titulo da figura")
```

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axis1.plot(x, y, 'g')
axis1.set_xlabel('Eixo X')
axis1.set_ylabel('Eixo Y')
axis1.set_title("titulo da figura 1")

axis2.plot(y, x, 'r')
axis2.set_xlabel('Eixo X')
axis2.set_ylabel('Eixo Y')
axis2.set_title("titulo da figura 2")
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'g-*')
ax.plot(x, x**4, 'r--')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("LOLLOLOL")
```

```python
#aula matplot 2
fig, ax = plt.subplots(figsize=(8, 10))

ax.plot(x*2, y, 'r-', label='teu cú')
ax.plot(x/3, y, 'g--', label='oloko')
ax.set_title('oi eu sou o goku!')
ax.legend(loc=0)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, x**2, 'y--')
ax.plot(x, x**3, 'g-*')
#use string spara alterar os estilos dos gráficos
#algumas strings que podem ser usadas são 'g=verde, y=amarelo, r=vermelho'
#outras são para alterar o estilo da linha como '-=tracejado, *=marcado com *'
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, x**2, color='#000000')
ax.plot(x, x**3, color='#FF0000')
#é possível alterar para qualquer cor usando a palete de cores do CSS
#exemplos: AliceBlue #F0F8FF, #FAEBD7 Aqua, #00FFFF Aquamarine, Black #000000
#para uma melhor noção, hexadecimal do CSS=RRGGBB, onde R=vermelho, G=green, B=blue
```

```python
#Tamanho da linha linewidth 
#transparência alpha
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, x**2, color='#000000', linewidth=10, alpha=0.5)
ax.plot(x, x**3, color='#FF0000', linewidth=35, alpha=0.1)
```

```python
#as possibilidades de layout para linhas

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# Possiveis estilos de linha: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Traços estilizados
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # Formato: comprimento da linha, comprimento do espaço, ...

# possíveis símbolos de marcador: marcador = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# tamanho e cor do marcador
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
#limitando x e y
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim([0,2])
ax.set_ylim([0,10])
ax.plot(x, x**2, color='#000000', linewidth=3, alpha=0.5)
ax.plot(x, x**3, color='#FF0000', linewidth=5, alpha=0.5)
```

```python
#note que devido a resolução do gráfico estar menor é possível
#visualizar as retas entre os pontos. Para suavisar as retas
#o banco de dados precisa ter mais pontos. ou seja um novo x se faz necessário.

#redefinindo x
x = np.linspace(0,5,50) #antes era 0,5,11
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim([0,2])
ax.set_ylim([0,10])
ax.plot(x, x**2, color='#000000', linewidth=3, alpha=0.5)
ax.plot(x, x**3, color='#FF0000', linewidth=5, alpha=0.5)
```

```python
#agora temos um gráfico com curvas, mas se limitarmos o x e o y
# veremos as retas novamente
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim([0,0.25])
ax.set_ylim([0,0.5])
ax.plot(x, x**2, color='#000000', linewidth=3, alpha=0.5)
ax.plot(x, x**3, color='#FF0000', linewidth=5, alpha=0.5)
```
