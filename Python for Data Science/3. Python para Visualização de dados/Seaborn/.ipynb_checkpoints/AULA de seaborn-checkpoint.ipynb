```python
#plots de distribuições

import seaborn as sns
```

```python
%matplotlib inline
#aqui é só pra ter sempre plotshow() sem precisar escrever
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
#aqui gera um distplot sem a linha de tendência
```

```python
sns.distplot(tips['total_bill'], bins=50)
#aqui gera um distplot com 50 barras
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') 
#com linha de regressão
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') 
#faz um heatmap
```

```python
sns.pairplot(tips)
#faz gráficos dos dados para todas as distribuição
#útil para melhor forma de visualização de dados
```

```python
sns.pairplot(tips, hue='sex')
#separa em duas cores os 'sex' dentro do dataset 'tips'
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
#gera cores para os dados maiores (azul) em relação aos menores (vermelho)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
#aula de plots categoricos
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
#aqui é baseado na média
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
#aqui é baseado no desvio padrão (np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.boxplot(data=tips, palette='rainbow',orient='h')
#somente para dados não quantitativos
```

```python
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips,  split=True, color='#000000')
sns.violinplot(x='day', y='total_bill', data=tips, jitter=True, split=True)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```
