```python
#plots de distribuições

import seaborn as sns
```

```python
%matplotlib inline
#aqui é só pra ter sempre plotshow() sem precisar escrever
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
#aqui gera um distplot sem a linha de tendência
```

```python
sns.distplot(tips['total_bill'], bins=50)
#aqui gera um distplot com 50 barras
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') 
#com linha de regressão
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') 
#faz um heatmap
```

```python
sns.pairplot(tips)
#faz gráficos dos dados para todas as distribuição
#útil para melhor forma de visualização de dados
```

```python
sns.pairplot(tips, hue='sex')
#separa em duas cores os 'sex' dentro do dataset 'tips'
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
#gera cores para os dados maiores (azul) em relação aos menores (vermelho)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
#aula de pair grids
import matplotlib.pyplot as plt
ir = sns.load_dataset('iris')
ir.head()
```

```python
ir['species'].value_counts()
```

```python
g= sns.PairGrid(ir)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(ir)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(ir, hue='species')
```

```python
#Retornou contagem qualitativa de fumantes em diferentes horários do dia
g1= sns.FacetGrid(tips, col='time', row='smoker')
g1.map(plt.hist, 'total_bill')
```

```python
sns.set_style('whitegrid') #darkgrid e white também existem
sns.countplot(x='sex', data=tips)
sns.despine(left=False, bottom=True, trim=True) #remover as bordas
```

```python
sns.countplot(x='sex', data=tips)
plt.figure(figsize= (12, 5))
```
