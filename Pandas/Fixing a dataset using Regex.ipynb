# Introduction

In this case, one will be transforming a dataset that was created with some problems.

```python
import pandas as pd
URL = 'https://caelum-online-public.s3.amazonaws.com/2925-introducao-pandas/banco_clientes.csv'
data = pd.read_csv(URL)
data
```

The problems of this dataset are Case in the nome columns, and () in the phone column.

To fix those problems one will put each name in the nome column in uppercase and remove all () from the phone column.

```python
data['nome'] = data['nome'].str.upper()
data.head()
```

```python
data['telefone'] = data['telefone'].str.replace('\(|\)','', regex=True)
data
```

## Splitting the address column

The endereco column has the address of each client. The address is currently composed of a single string and this could be splitted into Street, House Number and postal code.

```python
data['Postal_code'] = data['endereco'].str.split('(\d{5}-\d{3})', regex=True, expand=True)[1].str.strip()
data['Street'] = data['endereco'].str.replace('(\d{5}-\d{3})', '', regex=True).str.strip()
data['Street'] = data['Street'].str.replace(',','').str.strip()
data[['Street', 'Number']] =  data['Street'].str.split('\s(?=\d+$)', regex=True, expand=True)
data.drop('endereco', axis=1, inplace=True)
data
```

# Conclusions

The data is now transformed and ready to be used. The new pattern allows one to access each part of the client's address, along with the correct value of its phone number.
